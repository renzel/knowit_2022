```python
with open("encrypted.txt", "r") as f:
    julestjerne = f.read()
picture = ""

for line in julestjerne.split("\n"):
    row = ""

    for number in line.split(" "):
        if number == "":
            break
        if str(bin(int(number.strip()))).count("1") % 2 == 0:
            row += "  "
        else:
            row += "X "

    picture += row + "\n"
print(picture)
```
