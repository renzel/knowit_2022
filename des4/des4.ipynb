```python
def check_if_palindrom(number):
    str_number = str(number)
    if str_number == str(str_number)[::-1]:
        return True
    else:
        return False
total = 0
for i in range(1000):
    dec_pal = False
    oct_pal = False
    bin_pal = False
    if check_if_palindrom(i):
        dec_pal = True
    if check_if_palindrom(oct(i)[2:]):
        oct_pal = True
    if check_if_palindrom(bin(i)[2:]):
        bin_pal = True
    if dec_pal:
        total += i
    
print(total)
```
